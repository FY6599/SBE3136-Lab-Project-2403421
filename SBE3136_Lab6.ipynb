```python
import gradio as gr
from PIL import Image
from landingai.predict import Predictor
from landingai.visualize import overlay_predictions

# --- CONFIGURATION ---
ENDPOINT_ID = "8a3af3fc-1fe1-4d44-b49b-e7d83f074ae8"
API_KEY = "land_sk_M5NPpcoQiBjRbrNwghRVYIexahia7YqjcKxMQOqpuK7nXirwL9"

def landing_lens_inference(input_img):
    if input_img is None:
        return None, "Please upload an image."

    # 1. Initialize Predictor
    predictor = Predictor(ENDPOINT_ID, api_key=API_KEY)

    # 2. Run Inference
    # The SDK accepts PIL images directly
    predictions = predictor.predict(input_img)

    # 3. Create Visualization (Bounding Boxes/Labels)
    # This overlays the results on a copy of the original image
    output_img = overlay_predictions(predictions, input_img)

    # 4. Format text results for the user-friendly display
    if not predictions:
        result_text = "No objects detected."
    else:
        result_text = "### Predictions Found:\n"
        for pred in predictions:
            # Displays: Label (Confidence%)
            # Depending on model type, pred may have 'label_name' or 'label'
            label = getattr(pred, 'label_name', 'Object')
            score = getattr(pred, 'score', 0)
            result_text += f"- **{label}**: {score:.2%}\n"

    return output_img, result_text

# --- GRADIO INTERFACE ---
with gr.Blocks(title="LandingLens Object Detection") as demo:
    gr.Markdown("# LandingLens Image Inference")
    gr.Markdown("Upload an image to see predictions from your LandingLens model.")

    with gr.Row():
        with gr.Column():
            input_view = gr.Image(type="pil", label="Upload Image")
            submit_btn = gr.Button("Run Inference", variant="primary")

        with gr.Column():
            output_view = gr.Image(type="pil", label="Visual Result")
            output_details = gr.Markdown(label="Detection Details")

    submit_btn.click(
        fn=landing_lens_inference,
        inputs=input_view,
        outputs=[output_view, output_details]
    )

# Run with share=True for a public Colab link
demo.launch(share=True)
```
